## Run Stochastic Alternating Fair $k$-means algorithm

### initialize a MiniBatchFairKMeans object

```python
seed = 1
np.random.seed(seed)

data_name = 'Synthetic-equal'
fair_instance = MiniBatchFairKMeans(data_name, seed = 1, n_clu = 10) 
fair_instance.init_labels_centers()
```

```python
for i in range(fair_instance.n_demoGroups):
    print ("number of groups " + str(i), len(np.where(fair_instance.gender == i)[0])) # Male = 0

for k in range(fair_instance.n_clusters):
    print ("number of pts in cluster " + str(k), len(np.where(fair_instance.arr_init_labels[0, :] == k)[0]))
```

### plot demographic composition scatter

```python
colmap_demo = {0: 'k', 1: 'm', 2: 'g', 3:'y', 4:'DarkGreen', 5:'r', 6:'b', 7:'c', 8: 'LightBlue', 9:"Orange"}
colors = list(map(lambda x: colmap_demo[x], fair_instance.gender))

fig, ax1 = plt.subplots(figsize=(8, 6))
ax1.xaxis.set_tick_params(labelsize=22)
ax1.yaxis.set_tick_params(labelsize=22)
ax1.scatter(fair_instance.point_mat[:, 0], fair_instance.point_mat[:, 1], color =  colors, alpha = 0.5, edgecolor='k')
ax1.set_xlabel('$x_1$', fontsize = 22)
ax1.set_ylabel('$x_2$', fontsize = 22)
#     plt.title("Demograpic composition: black->males, purple->females")

plt.tight_layout()
plt.show()
```

### run PF-SAfairKM algorithm to get Pareto front approximation

```python
fair_instance.n_init = 30 ## number of initial clustering labels
fair_instance.max_iter = 400 ## maximum number of iterations to run
fair_instance.point_per_iteration = 1 ## number of new point generated from any current point with a pair of (n_a, n_b)
fair_instance.max_len_pareto_front = 1500 ## maximum number of points on Pareto front
fair_instance.dense_threshold = 1/400.0 #1/500 for bank, 1/300 for Adult
```

```python
f1_value_list, f2_value_list, updating_center_list, updating_label_list, \
total_time = fair_instance.main_PFSAD(seed = seed)
```

```python
## dump data into pickle file
# logfile = "pickle/PFSAG_%s_normalized_iters%s_numpts%s.pickle"%(data_name, fair_instance.num_iter, len(f1_value_list))
# pickle.dump([f1_value_list, f2_value_list, updating_center_list, updating_label_list, total_time], open(logfile,"wb"))
```

### plot the obtained Pareto front

```python
def remove_pts(list_f1, list_f2, list_centers, list_labels, num):
    list_f1_idx = np.argsort(list_f1)[:-num]
    return list_f1[list_f1_idx], list_f2[list_f1_idx], list_centers[list_f1_idx], list_labels[list_f1_idx]
```

```python
f1_value_list, f2_value_list, updating_center_list, updating_label_list = \
remove_pts(f1_value_list, f2_value_list, updating_center_list, updating_label_list, 1)
```

```python
fig = plt.figure(figsize=(12, 9), dpi= 80, facecolor='w', edgecolor='k')
ax1 = plt.subplot(1, 1, 1)
ax1.xaxis.set_tick_params(labelsize=22)
ax1.yaxis.set_tick_params(labelsize=22)
ax1.plot(f1_value_list, f2_value_list, 'bo', lw=2, alpha = 1)
ax1.set_xlabel('$f_1(s)$', fontsize = 22)
ax1.set_ylabel('$f_2(s)$', fontsize = 22)
# plt.title("Nondominated points after second cleaning")

plt.tight_layout()
plt.show()
```
